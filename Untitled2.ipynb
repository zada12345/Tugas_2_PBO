```python
import math

class Lingkaran:
    def __init__(self, jari_jari):
        self.jari_jari = jari_jari

    def hitung_luas(self):
        return math.pi * self.jari_jari ** 2

# Membuat objek
blt = Lingkaran(7)
print(f"Luas lingkaran: {blt.hitung_luas():.2f}")
```

```python
class Persegi:
    def __init__(self, sisi):
        self.sisi = sisi

    def hitung_luas(self):
        return self.sisi ** 2

# Membuat objek
sq = Persegi(4)
print(f"Luas persegi: {sq.hitung_luas()}")
```

```python
class Mobil:
    def __init__(self, merek, warna):
        self.merek = merek
        self.warna = warna

    def info(self):
        print(f"Mobil {self.merek} berwarna {self.warna}.")

# Membuat objek
avanza = Mobil("Toyota Avanza", "Hitam")
avanza.info()
```

```python
class Pegawai:
    def __init__(self, nama, gaji):
        self.nama = nama
        self.gaji = gaji

    def info_pegawai(self):
        print(f"Nama: {self.nama}, Gaji: Rp{self.gaji:,}")

# Membuat objek
pgw = Pegawai("Andi", 5000000)
pgw.info_pegawai()
```

```python
class Mahasiswa:
    def __init__(self, nama, nim):
        self.nama = nama
        self.nim = nim

    def cetak_krs(self):
        print(f"Mahasiswa {self.nama} (NIM: {self.nim}) telah mencetak KRS.")

# Membuat objek
mhs = Mahasiswa("Budi Santoso", "210001")
mhs.cetak_krs()
```
